### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
def generate_random_matrix(m=3, n=3, min_limit=0, max_limit=10):
    matrix = []
    for i in range(m):
        row = []
        for j in range(n):
            row.append(random.randint(min_limit, max_limit))
        matrix.append(row)
    return matrix
random_matrix = generate_random_matrix(m=100, n=100, min_limit=-1500, max_limit=1500)
print(random_matrix)
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
def selection_sort(array):
    for i in range(len(array)-1):
        min_idx = i
        for idx in range(i + 1, len(array)):
            if array[idx] < array[min_idx]:
                min_idx = idx
        array[i], array[min_idx] = array[min_idx], array[i]
    return array
for i in range(len(random_matrix)):
    selection_sort(random_matrix[i])
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print(random_matrix)
start_time = time.time()
for i in range(len(random_matrix)):
    random_matrix[i].sort()
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
start_time = time.time()
def insertion_sort(array):
    n = len(array)
    for i in range(1, n):
        x = array[i]
        j = i
        while j > 0 and array[j - 1] > x:
            array[j] = array[j - 1]
            j -= 1
        array[j] = x
    return array
for i in range(len(random_matrix)):
    insertion_sort(random_matrix[i])
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print(random_matrix)
start_time = time.time()
for i in range(len(random_matrix)):
    random_matrix[i].sort()
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. | Сортировка пузырьком.
start_time = time.time()
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
for i in range(len(random_matrix)):
    bubble_sort(random_matrix[i])
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print(random_matrix)
start_time = time.time()
for i in range(len(random_matrix)):
    random_matrix[i].sort()
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()
def shell_sort(arr):
    n = len(arr)
    gap = n // 2

    while gap > 0:
        for i in range(gap, n):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = temp
        gap //= 2

    return arr
for i in range(len(random_matrix)):
    shell_sort(random_matrix[i])
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print(random_matrix)
start_time = time.time()
for i in range(len(random_matrix)):
    random_matrix[i].sort()
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
start_time = time.time()
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less = [x for x in arr[1:] if x <= pivot]
        greater = [x for x in arr[1:] if x > pivot]
        return quicksort(less) + [pivot] + quicksort(greater)
for i in range(len(random_matrix)):
    quicksort(random_matrix[i])
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print(random_matrix)
start_time = time.time()
for i in range(len(random_matrix)):
    random_matrix[i].sort()
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.
start_time = time.time()
def tournament_sort(array):
    def merge(left, right):
        result = []
        while left and right:
            if left[0] < right[0]:
                result.append(left.pop(0))
            else:
                result.append(right.pop(0))
        result.extend(left)
        result.extend(right)
        return result

    def sort(arr):
        if len(arr) <= 1:
            return arr
        mid = len(arr) // 2
        left = sort(arr[:mid])
        right = sort(arr[mid:])
        return merge(left, right)

    return sort(array)
for i in range(len(random_matrix)):
    tournament_sort(random_matrix[i])
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
print(random_matrix)
start_time = time.time()
for i in range(len(random_matrix)):
    random_matrix[i].sort()
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
